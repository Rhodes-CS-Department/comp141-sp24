# Day 22 Using Loops to Process Files
* Files typically used to hold large amounts of data, so it is often best to use a loop to process a file
* Often the number of items stored in file is unknown
* The `readline` method uses an empty string as a sentinel when end of file is reached
```
line = file.readline()
while line != '':
    #do something with the line
    line = file.readline()
```


## Steps for Processing a File
1. Open the file
2. Loop over the file
3. Process each line
4. Close the file



````python
# This program reads all the values from a file called sales.txt
# and keeps a running total of all the values in the file

def main():
    # this uses the open function, the first argument is the file name,
    # the second argument is the mode (read = "r", write = "w")
    # sales_file will be File object (ie, has type File)
    sales_file = open("sales.txt", 'r') 
    
    total = 0
    #Read the first line of the file, but don't convert it to a number until we test it to be sure it's not empty
    line = sales_file.readline() # reads the next line
    line = sales_file.readline()
    
    while line != '':
        #Convert line to a float
        amount = float(line) # assumes the line will have a number in it
    
        #formats and displays the amount, then adds it to total
        print(format(amount, '.2f'))
        total += amount
        
    
        #Read the next line in the file
        line = sales_file.readline()

    #Don't forget to close the file
    sales_file.close()
  
    print("Your total sales is $", format(total, ',.2f'))

main()
````

Using a while loop to process a file helps to show the intuition behind how it works. However, it is __much__ easier to process files using for loops. Python has a special syntax available for doing this.

## Using Python’s for Loop to Read Lines
* The for loop will automatically read lines from a file, and stop when it reaches the end of the file
* No longer need to use the `readline` method or check to make sure `line != ''`
* The loop iterates once over each line in the file
* Syntax:
```
for line in file_object:
    statements
```

````python
# This program uses a for loop to read all the values from a file called sales.txt
# and keeps a running total of all the values in the file

def main():
    # sales_file is the File object
    sales_file = open("sales.txt", 'r')
    
    total = 0
    headers = sales_file.readline()
    headers = headers.rstrip() # rstrip will strip blank/newline characters from the right
    print(headers)
  
    # since this for loop will do our calculation, we do want to manually
    # skip the header
    for line in sales_file:
        #Convert line to a float
        amount = float(line)
    
        #formats and displays the amount, then adds it to total
        print(format(amount, '.2f'))
        total += amount

    #Don't forget to close the file
    sales_file.close()
  
    print("Your total sales is $", format(total, ',.2f'))

main()
````

## Processing each line of your file
* It's usually a good idea to 'strip' the newline character from the line before processing it (especially for strings)
* When processing a number, converting the string to a number will automatically remove the whitespace
* A good programming practice is to `print` the file contents as you process the file. This will help with debugging your code.

````python
import random

# make a file containing a few random numbers
n = 10 # number of random numbers in my list
a = 0 # lower bound for the random numbers
b = 1000 # upper bound for the random numbers

# make the file that we will store the random numbers
random_numbers_file = open("randomNums.txt", "w")

for i in range(n):
    new_random_number = random.randint(a, b)
    random_numbers_file.write(str(new_random_number))
    random_numbers_file.write("\n") # \ is called an escape character, it lets you talk about special characters
    
random_numbers_file.close()
````

````python
# example of the sliding window technique:
# find the consecutive difference between these numbers
# uses a variable to track a previous value and update the previous value as we go
numbers = [1, 3, 10, 12, 9, 6, 4]
prev_num = 1
for i in range(1, len(numbers)):
    current_number = numbers[i] # gives the current number at index i
    consecutive_difference = current_number - prev_num
    prev_num = current_number
    print(consecutive_difference)
````
