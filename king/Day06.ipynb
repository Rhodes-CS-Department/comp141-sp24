## When if-else is not enough





````python
#This program gets a numeric test score from the user and displays the corresponding letter grade

#Constants for the grade thresholds
A_SCORE = 90
B_SCORE = 80
C_SCORE = 70
D_SCORE = 60

#Get the score from the user
score = int(input("Enter your test score: "))

#Determine the correct letter grade
if score >= A_SCORE:
    print('Your grade is an A.')
else:
    if score >= B_SCORE:
        print('Your grade is a B.')
    else:
        if score >= C_SCORE:
            print('Your grade is a C.')
        else:
            if score >= D_SCORE:
                print('Your grade is a D.')
            else:
                print('Your grade is an F.')
      
````

###### The if-elif-else statement

__if-elif-else statement:__ special version of a decision structure
* Makes logic of nested decision structures simpler to write
* Can include multiple elif statements
* Syntax: 	
```
            if condition1:
				statements
			elif condition2:
				statements 
			else:
				statements
```



````python
#This program gets a numeric test score from the user and displays the corresponding letter grade
#This program is equivalent to the previous program, but is much more readable

#Constants for the grade thresholds
A_SCORE = 90
B_SCORE = 80
C_SCORE = 70
D_SCORE = 60

#Get the score from the user
score = int(input("Enter your test score: "))

#Determine the correct letter grade

if score >= A_SCORE:
    print('Your grade is an A.')
elif score >= B_SCORE:
    print('Your grade is a B.')
elif score >= C_SCORE:
    print("Your grade is a C.")
elif score >= D_SCORE:
    print('Your grade is a D.')
else:
    print('Your grade is an F.')
````

## Logical Operators

__Logical operators:__ operators that can be used to create complex Boolean expressions
* `and` operator and `or` operator: binary operators, connect two Boolean expressions into a compound Boolean expression
* `not` operator: unary operator, reverses the truth of its Boolean operand

__The `and` operator__: Both individual tests must be True to make the entire statement True.

```
if ___________ and ______________:
    #do something
    
else:
    #do something else
```

Truth Table for `and` operator

Expression | Value of Expression
------------ | -------------
False and False | False
False and True | False
True and False | False
True and True | True

__The `or` operator__: Both individual tests must be False to make the entire statement False.

```
if ___________ or ______________:
    #do something
    
else:
    #do something else
```

Truth Table for `or` operator

Expression | Value of Expression
------------ | -------------
False and False | False
False and True | True
True and False | True
True and True | True


__The `not` operator__: Takes a Boolean expression as operand and reverses its logical value
* Sometimes it may be necessary to place parentheses around an expression to clarify to what you are applying the not operator
```
if not (__________):
    #do something
```

Truth Table for `not` operator

Expression | Value of Expression
------------ | -------------
False | True
True | False

## Checking Numeric Ranges with Logical Operators

To determine whether a numeric value is within a specific range of values, use `and` 
* Example: `x >= 10 and x <= 20`

To determine whether a numeric value is outside of a specific range of values, use `or` 
* Example: `x < 10 or x > 20`


## Short Circuit Evaluation

__Short circuit evaluation:__ deciding the value of a compound Boolean expression after evaluating only one sub expression
* Performed by the or and and operators
* For or operator: If left operand is true, compound expression is true. Otherwise, evaluate right operand
* For and operator: If left operand is false, compound expression is false. Otherwise, evaluate right operand


````python
#Examples of Short Circuit Evaluation with and operator

x = int(input()) #Try with 6, 4, and 12
if x >= 5 and x <= 10:
    print("Here")
````

````python
#Examples of Short Circuit Evaluation with or operator

x = int(input()) #Try with 10, 45, 66
if x <= 10 or x >= 65:
    print("here")
````

## Nested Decision Structures

Sometimes the way information is presented, it seems like a nested if/else statement is necessary.

Take for example this flow chart for a loan qualification program.





````python
#This program determines whether a bank customer qualifies for a loan.

#Constants for minimum salary and minimum years on the job
MIN_SALARY = 30000.0
MIN_YEARS = 2

#Get the customer's annual salary
salary = float(input("Enter your annual salary: "))

#Get the number of years on the current job
years_on_job = int(input("Enter the number of years employed: "))

#Determine whether the customer qualifies
if salary >= MIN_SALARY:
    if years_on_job >= MIN_YEARS:
        print("You qualify for the loan.")
    else:
        print("You must have been employed for at least", MIN_YEARS, "years to qualify.")
else:
    print("You must earn at least $", format(MIN_SALARY, '.2f'), " per year to qualify.", sep='')
````
